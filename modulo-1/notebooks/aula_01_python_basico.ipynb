## 2. Variáveis e Tipos de Dados

Uma **variável** é um nome que aponta para um valor na memória.
Python é **dinamicamente tipado** — o tipo é inferido automaticamente.

| Tipo | Exemplo | Descrição |
|------|---------|------------|
| `int` | `42` | Números inteiros |
| `float` | `3.14` | Números decimais |
| `str` | `"texto"` | Cadeias de caracteres |
| `bool` | `True` / `False` | Verdadeiro ou Falso |
| `NoneType` | `None` | Ausência de valor |

```python
# ----- Inteiros (int) -----
idade = 28
ano_nascimento = 1996
print(type(idade))          # <class 'int'>

# ----- Decimais (float) -----
altura = 1.75
salario = 4500.50
print(type(altura))         # <class 'float'>

# ----- Texto (str) -----
nome = "Ana Silva"
cidade = 'Fortaleza'        # aspas simples ou duplas são equivalentes
print(type(nome))           # <class 'str'>

# ----- Booleano (bool) -----
esta_matriculado = True
pagou_mensalidade = False
print(type(esta_matriculado))  # <class 'bool'>

# ----- Nenhum valor (None) -----
resultado = None
print(type(resultado))      # <class 'NoneType'>
```

```python
# ========== OPERAÇÕES ARITMÉTICAS ==========

a = 10
b = 3

print(f"Soma:          {a} + {b} = {a + b}")
print(f"Subtração:     {a} - {b} = {a - b}")
print(f"Multiplicação: {a} * {b} = {a * b}")
print(f"Divisão:       {a} / {b} = {a / b:.4f}")   # divisão sempre retorna float
print(f"Divisão int:   {a} // {b} = {a // b}")      # parte inteira da divisão
print(f"Módulo (resto):{a} % {b} = {a % b}")
print(f"Potência:      {a} ** {b} = {a ** b}")
```

```python
# ========== STRINGS EM DETALHES ==========

texto = "Python para Ciência de Dados"

# Comprimento da string
print(f"Tamanho: {len(texto)} caracteres")

# Fatiamento (slicing): texto[início:fim:passo]
print(f"Primeiros 6 chars: {texto[:6]}")
print(f"Últimos 5 chars:   {texto[-5:]}")
print(f"De índice 7 a 10:  {texto[7:10]}")

# Métodos úteis de string
print(texto.upper())          # TUDO EM MAIÚSCULO
print(texto.lower())          # tudo em minúsculo
print(texto.replace("Dados", "Analytics"))  # substituir
print(texto.split(" "))       # dividir em lista de palavras

# F-string (forma moderna de formatar strings — Python 3.6+)
nome = "Carlos"
nota = 9.75
print(f"Aluno: {nome} | Nota: {nota:.1f}")
```

```python
# ========== CONVERSÃO DE TIPOS (CASTING) ==========

# str → int
numero_texto = "42"
numero_int = int(numero_texto)
print(f"'{numero_texto}' como int: {numero_int}", type(numero_int))

# int → float
inteiro = 7
decimal = float(inteiro)
print(f"{inteiro} como float: {decimal}")

# float → int  ⚠️ trunca a parte decimal (não arredonda!)
pi = 3.99
print(f"{pi} como int: {int(pi)}")   # resultado: 3, não 4!

# número → str
valor = 1500
mensagem = "Saldo: R$ " + str(valor)
print(mensagem)
```

## 3. Estruturas Condicionais

Usadas para **tomar decisões** no código com base em condições.
A indentação (4 espaços) é **obrigatória** em Python — ela define os blocos de código.

```python
# ========== IF / ELIF / ELSE ==========

nota = 7.5

if nota >= 9.0:
    conceito = "A"
    print(f"Nota {nota} → Conceito {conceito} — Excelente! 🏆")
elif nota >= 7.0:
    conceito = "B"
    print(f"Nota {nota} → Conceito {conceito} — Bom! 👍")
elif nota >= 5.0:
    conceito = "C"
    print(f"Nota {nota} → Conceito {conceito} — Regular 📚")
else:
    conceito = "D"
    print(f"Nota {nota} → Conceito {conceito} — Reprovado ❌")
```

```python
# ========== OPERADORES LÓGICOS ==========
# and → ambas as condições verdadeiras
# or  → pelo menos uma condição verdadeira
# not → inverte o valor lógico

idade = 20
tem_cnh = True

if idade >= 18 and tem_cnh:
    print("Pode dirigir ✅")
else:
    print("Não pode dirigir ❌")

# Operadores de comparação
x = 10
print(x == 10)   # igual
print(x != 5)    # diferente
print(x > 8)     # maior que
print(x <= 10)   # menor ou igual

# Operador ternário (if em linha única)
status = "Maior de idade" if idade >= 18 else "Menor de idade"
print(status)
```

## 4. Estruturas de Repetição

Permitem **repetir** blocos de código múltiplas vezes.
- `for` → quando sabemos o número de repetições (itera sobre uma sequência)
- `while` → quando repetimos enquanto uma condição for verdadeira

```python
# ========== FOR LOOP ==========

# Iterando sobre uma lista
frutas = ["maçã", "banana", "uva", "manga"]
for fruta in frutas:
    print(f"Fruta: {fruta}")

print("---")

# range(início, fim, passo) — gera uma sequência de números
# range(5) → 0, 1, 2, 3, 4  (fim é exclusivo!)
for i in range(1, 6):
    print(f"{i} x 3 = {i * 3}")

print("---")

# enumerate() → retorna índice + valor
for indice, fruta in enumerate(frutas, start=1):
    print(f"{indice}. {fruta}")
```

```python
# ========== WHILE LOOP ==========

# Conta regressiva — repete enquanto contador > 0
contador = 5
while contador > 0:
    print(f"Contagem: {contador}")
    contador -= 1   # equivale a: contador = contador - 1
print("🚀 Lançamento!")

print("---")

# break → interrompe o loop antes do fim
# continue → pula para a próxima iteração

numeros = [1, 2, 3, -1, 5, 6]
for num in numeros:
    if num < 0:
        print(f"Número negativo encontrado: {num}. Parando.")
        break             # para o loop ao encontrar negativo
    print(f"Processando: {num}")
```

```python
# ========== LIST COMPREHENSION ==========
# Forma concisa e pythônica de criar listas
# Sintaxe: [expressão for item in iterável if condição]

# Forma tradicional
quadrados_trad = []
for n in range(1, 11):
    quadrados_trad.append(n ** 2)

# Forma com list comprehension (equivalente, mas mais elegante)
quadrados = [n ** 2 for n in range(1, 11)]
print("Quadrados:", quadrados)

# Com condição: apenas os pares
pares = [n for n in range(1, 21) if n % 2 == 0]
print("Pares:", pares)

# Transformar strings
nomes = ["ana", "bruno", "carla"]
nomes_maiusc = [nome.capitalize() for nome in nomes]
print("Nomes:", nomes_maiusc)
```

## 5. Estruturas de Dados

| Estrutura | Sintaxe | Ordenada | Mutável | Duplicatas |
|-----------|---------|----------|---------|------------|
| `list` | `[1, 2, 3]` | ✅ | ✅ | ✅ |
| `tuple` | `(1, 2, 3)` | ✅ | ❌ | ✅ |
| `dict` | `{"a": 1}` | ✅ (3.7+) | ✅ | ❌ (chaves) |
| `set` | `{1, 2, 3}` | ❌ | ✅ | ❌ |

```python
# ========== LISTAS ==========
# Coleção ordenada e mutável. Aceita tipos mistos.

notas = [8.5, 7.0, 9.2, 6.8, 10.0]

# Acessando por índice (começa no 0)
print(f"Primeira nota: {notas[0]}")
print(f"Última nota:   {notas[-1]}")
print(f"Notas 2 a 4:   {notas[1:4]}")

# Métodos principais
notas.append(8.0)           # adiciona ao final
notas.insert(0, 7.5)        # insere no índice 0
notas.remove(6.8)           # remove primeiro elemento igual a 6.8
removido = notas.pop()      # remove e retorna o último elemento

print(f"Lista final: {notas}")
print(f"Tamanho: {len(notas)}")
print(f"Maior nota: {max(notas)} | Menor: {min(notas)} | Média: {sum(notas)/len(notas):.2f}")

# Ordenação
notas.sort()                # ordena no lugar (modifica a lista)
print(f"Ordenada:    {notas}")
notas.sort(reverse=True)
print(f"Decrescente: {notas}")
```

```python
# ========== TUPLAS ==========
# Coleção ordenada e IMUTÁVEL. Ideal para dados que não devem mudar.

coordenadas = (-3.7172, -38.5433)   # latitude e longitude de Fortaleza
print(f"Latitude: {coordenadas[0]} | Longitude: {coordenadas[1]}")

# Unpacking (desempacotamento)
lat, lon = coordenadas
print(f"Desempacotado → lat: {lat}, lon: {lon}")

# Tuplas são mais rápidas que listas para leitura
cores_rgb = (255, 0, 0)   # vermelho — não faz sentido mudar isso
print(f"Cor RGB: {cores_rgb}")

# Não é possível modificar:
# coordenadas[0] = 0  → TypeError!
```

```python
# ========== DICIONÁRIOS ==========
# Coleção de pares chave:valor. Chaves devem ser únicas.

aluno = {
    "nome": "Maria Oliveira",
    "idade": 22,
    "curso": "Ciência de Dados",
    "notas": [8.5, 9.0, 7.8],
    "ativo": True
}

# Acessando valores
print(aluno["nome"])                    # acesso direto (KeyError se não existir)
print(aluno.get("email", "N/A"))        # acesso seguro com valor padrão

# Adicionando e modificando
aluno["email"] = "maria@email.com"      # adiciona nova chave
aluno["idade"] = 23                     # modifica valor existente

# Iterando
print("\n--- Dados do Aluno ---")
for chave, valor in aluno.items():
    print(f"  {chave:10}: {valor}")

# Verificando existência de chave
if "notas" in aluno:
    media = sum(aluno["notas"]) / len(aluno["notas"])
    print(f"\nMédia: {media:.2f}")
```

```python
# ========== SETS (CONJUNTOS) ==========
# Coleção não ordenada, sem duplicatas. Ótima para operações de conjunto.

linguagens_turma_a = {"Python", "R", "SQL", "Python", "Java"}
linguagens_turma_b = {"Python", "SQL", "Scala", "Julia"}

print(f"Turma A: {linguagens_turma_a}")   # Python aparece só uma vez
print(f"Turma B: {linguagens_turma_b}")

# Operações de conjunto
print(f"\nUnião (A ∪ B):        {linguagens_turma_a | linguagens_turma_b}")
print(f"Interseção (A ∩ B):   {linguagens_turma_a & linguagens_turma_b}")
print(f"Diferença (A - B):    {linguagens_turma_a - linguagens_turma_b}")
print(f"Diferença simétrica:  {linguagens_turma_a ^ linguagens_turma_b}")
```

## 6. Boas Práticas de Programação

Escrever código que funciona é o mínimo. Escrever código **legível, organizado e sustentável** é o objetivo.

### Principais guias
- **PEP 8**: guia de estilo oficial do Python
- **PEP 257**: convenções para docstrings
- **DRY**: *Don't Repeat Yourself* — evite duplicação de código
- **KISS**: *Keep It Simple, Stupid* — prefira soluções simples

```python
# ========== NOMES DESCRITIVOS ==========

# ❌ Ruim — nomes sem significado
x = [8.5, 7.0, 9.2]
t = sum(x) / len(x)

# ✅ Bom — nomes que explicam o propósito
notas_aluno = [8.5, 7.0, 9.2]
media_final = sum(notas_aluno) / len(notas_aluno)
print(f"Média final: {media_final:.2f}")

# Convenções de nomenclatura:
# snake_case     → variáveis e funções: nome_completo, calcular_media
# PascalCase     → classes: MinhaClasse, ProcessadorDados
# UPPER_SNAKE    → constantes: TAXA_JUROS, PI
# _prefixo       → privado/interno: _contador_interno

TAXA_DESCONTO = 0.10      # constante — não deve ser alterada
PI = 3.14159
```

```python
# ========== FUNÇÕES REUTILIZÁVEIS ==========
# Princípio DRY: encapsule lógica repetida em funções

def calcular_media(lista_notas: list) -> float:
    """
    Calcula a média aritmética de uma lista de notas.

    Parâmetros:
        lista_notas (list): Lista com valores numéricos.

    Retorna:
        float: Média dos valores. Retorna 0.0 se a lista estiver vazia.

    Exemplo:
        >>> calcular_media([8.0, 9.0, 7.0])
        8.0
    """
    if not lista_notas:           # lista vazia
        return 0.0
    return sum(lista_notas) / len(lista_notas)


def classificar_aluno(media: float) -> str:
    """Retorna a classificação do aluno com base na média."""
    if media >= 9.0:
        return "Aprovado com Distinção 🏆"
    elif media >= 7.0:
        return "Aprovado ✅"
    elif media >= 5.0:
        return "Recuperação ⚠️"
    else:
        return "Reprovado ❌"


# Usando as funções
alunos = {
    "João": [8.5, 9.0, 7.8, 8.2],
    "Carla": [6.0, 5.5, 7.0, 6.8],
    "Pedro": [4.0, 3.5, 5.0, 4.8],
}

print("=" * 40)
print(f"{'Aluno':<10} {'Média':>6}  {'Status'}")
print("=" * 40)
for nome, notas in alunos.items():
    media = calcular_media(notas)
    status = classificar_aluno(media)
    print(f"{nome:<10} {media:>6.2f}  {status}")
```

```python
# ========== TRATAMENTO DE ERROS ==========
# Use try/except para lidar com erros esperados sem travar o programa

def dividir(a: float, b: float) -> float:
    """Divide dois números com tratamento de erro para divisão por zero."""
    try:
        resultado = a / b
        return resultado
    except ZeroDivisionError:
        print("⚠️  Erro: divisão por zero não é permitida.")
        return None
    except TypeError as e:
        print(f"⚠️  Erro de tipo: {e}")
        return None

print(dividir(10, 2))      # 5.0
print(dividir(10, 0))      # mensagem de erro + None
print(dividir(10, "a"))    # mensagem de erro + None

# Usando assert para validações (útil em desenvolvimento)
def calcular_imc(peso_kg: float, altura_m: float) -> float:
    """Calcula o Índice de Massa Corporal."""
    assert peso_kg > 0, "Peso deve ser positivo"
    assert 0.5 < altura_m < 3.0, "Altura deve estar entre 0.5m e 3.0m"
    return peso_kg / (altura_m ** 2)

imc = calcular_imc(70, 1.75)
print(f"\nIMC: {imc:.1f}")
```
