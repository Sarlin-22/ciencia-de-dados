## 1. Primeiro contato com Python

Python é uma linguagem de programação bastante utilizada em diferentes áreas, incluindo automação, desenvolvimento de sistemas e Ciência de Dados.

Vamos começar com um comando simples de saída.

```python
print("Olá, mundo!")
```

## 2. Variáveis

Variáveis são usadas para armazenar informações.  
Elas podem guardar textos, números e outros tipos de dados.

```python
nome = "Maria"
idade = 20
altura = 1.65

print(nome)
print(idade)
print(altura)
```

## 3. Tipos de dados

Alguns tipos básicos em Python:

- `str`: textos
- `int`: números inteiros
- `float`: números decimais
- `bool`: verdadeiro ou falso

```python
nome = "Carlos"
idade = 18
nota = 8.5
aprovado = True

print(type(nome))
print(type(idade))
print(type(nota))
print(type(aprovado))
```

## 4. Operações básicas

Python permite realizar operações matemáticas de forma simples.

```python
a = 10
b = 5

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
```

## 5. Entrada de dados

Também é possível receber valores digitados pelo usuário usando `input()`.

No exemplo abaixo, o valor digitado é tratado como texto.  
Quando queremos trabalhar com números, precisamos converter esse valor.

```python
nome = input("Digite seu nome: ")
print("Olá,", nome)
```

```python
numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

soma = numero1 + numero2
print("Resultado da soma:", soma)
```

## 6. Estruturas condicionais

As estruturas condicionais permitem tomar decisões no programa.

Principais comandos:
- `if`
- `elif`
- `else`

```python
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

## 7. Estruturas de repetição

As repetições servem para executar comandos várias vezes.

Neste módulo, vamos trabalhar com:
- `for`
- `while`

```python
for i in range(1, 6):
    print("Número:", i)
```

```python
contador = 1

while contador <= 5:
    print("Contador:", contador)
    contador += 1
```

## 8. Listas

Listas permitem armazenar vários valores em uma única variável.

```python
frutas = ["maçã", "banana", "uva", "laranja"]

print(frutas)
print("Primeira fruta:", frutas[0])
print("Última fruta:", frutas[-1])
```

```python
frutas.append("abacaxi")
print(frutas)
```

## 9. Tuplas

Tuplas também armazenam múltiplos valores, mas não podem ser alteradas após a criação.

```python
dias_semana = ("segunda", "terça", "quarta", "quinta", "sexta")

print(dias_semana)
print("Terceiro dia:", dias_semana[2])
```

## 10. Dicionários

Dicionários armazenam informações em formato de chave e valor.

```python
aluno = {
    "nome": "João",
    "idade": 19,
    "curso": "Python"
}

print(aluno)
print("Nome:", aluno["nome"])
print("Idade:", aluno["idade"])
print("Curso:", aluno["curso"])
```
