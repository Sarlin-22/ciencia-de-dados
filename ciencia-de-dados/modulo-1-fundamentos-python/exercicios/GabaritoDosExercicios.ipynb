```python
# Exercício 2
nome = "Rilder"
print("Olá,", nome + "! Seja bem-vindo ao curso.")

# Outra forma:

nome = "Rilder"
print(f"Olá, {nome}! Seja bem-vindo ao curso.")
```

```python
# Exercício 3
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))

print("Soma:", num1 + num2)
print("Subtração:", num1 - num2)
print("Multiplicação:", num1 * num2)
print("Divisão:", num1 / num2)
```

```python
# Exercício 4
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

```python
# Exercício 5
numero = float(input("Digite um número: "))

if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Igual a zero")
```

```python
# Exercício 6
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
# Exercício 7
for i in range(1, 11):
    print(i)
```

```python
# Exercício 8
soma = 0

for i in range(1, 6):
    soma = soma + i

print("Soma:", soma)
```

```python
# Exercício 9
frutas = ["maçã", "banana", "uva", "laranja", "abacaxi"]

print("Lista completa:", frutas)
print("Primeira fruta:", frutas[0])
print("Última fruta:", frutas[-1])
```

```python
# Exercício 10
numeros = [2, 4, 6, 8, 10]

for numero in numeros:
    print(numero * 2)
```

```python
# Exercício 11
dias_semana = ("segunda", "terça", "quarta", "quinta", "sexta", "sábado", "domingo")

print("Tupla completa:", dias_semana)
print("Terceiro dia:", dias_semana[2])
```
