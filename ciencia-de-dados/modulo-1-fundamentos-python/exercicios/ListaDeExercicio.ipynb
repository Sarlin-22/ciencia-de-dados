# Lista de Exercícios - Fundamentos de Python

# Exercício 1
Crie um programa que exiba na tela a frase:

Olá, mundo!

# Exercício 2
Crie um programa que armazene seu nome em uma variável e depois mostre a seguinte mensagem:

Olá, [nome]! Seja bem-vindo ao curso.

# Exercício 3
Crie um programa que receba dois números inteiros e mostre:

a soma;
a subtração;
a multiplicação;
a divisão.

# Exercício 4
Faça um programa que receba a idade de uma pessoa e informe se ela é:

menor de idade;
maior de idade. 

Considere maior de idade quem tem 18 anos ou mais.

# Exercício 5
Crie um programa que receba um número e informe se ele é:

positivo;
negativo;
igual a zero.

# Exercício 6
Faça um programa que receba uma nota de 0 a 10 e informe se o aluno está:

aprovado, se a nota for maior ou igual a 7;
reprovado, se a nota for menor que 7.

# Exercício 7
Use um laço de repetição para mostrar na tela os números de 1 até 10.

# Exercício 8
Faça um programa que mostre na tela a soma dos números de 1 até 5.

# Exercício 9
Crie uma lista com 5 frutas e:

mostre a lista completa;
mostre apenas o primeiro elemento;
mostre apenas o último elemento.

# Exercício 10
Crie um programa que percorra uma lista de números [2, 4, 6, 8, 10] e mostre cada número multiplicado por 2.

# Exercício 11
Crie uma tupla com os dias da semana e mostre:

a tupla completa;
o terceiro dia da semana.



# Exercício 12
Crie um dicionário com as seguintes informações de uma pessoa:

nome
idade
cidade

Depois, mostre na tela o valor de cada uma dessas informações.
